```python
# just to print the base information of the data set


import pandas as pd

movie_genres_csv_path = "./data/movie_genres.csv"
user_reviews_csv_path = "./data/user_reviews.csv"

movie_genres = pd.read_csv(movie_genres_csv_path)
user_reviews = pd.read_csv(user_reviews_csv_path)

```

```python
movie_shape = movie_genres.shape
user_shape = user_reviews.shape
print("movie shape:", movie_shape)
print(movie_genres.columns)

print("user shape:", user_shape)
print(user_reviews.columns)
```

```python
user_reviews = user_reviews.iloc[:,2:]
non_zero_count = user_reviews.apply(lambda x: x[x != 0].count())
print(non_zero_count)

import matplotlib.pyplot as plt 
plt.figure(figsize=(10, 6))
non_zero_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('')
plt.title('how many person rated for each movies')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
movie_genres = movie_genres.iloc[:,2:]
non_zero_count = movie_genres.apply(lambda x: x[x != 0].count(), axis=1)
print(non_zero_count)

import matplotlib.pyplot as plt 
plt.figure(figsize=(10, 6))
non_zero_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('')
plt.title('how many movies rated by each person')
plt.xticks(rotation=45, ha='right')
plt.show()
```
